```python
import pandas as pd
```

```python
transactions = pd.read_csv('./raw_transactions.csv')
transactions
```

```python
exactInputSingleTransactions = transactions[transactions['Tx_decoded_function'] == 'exactInputSingle']
exactInputSingleTransactions['Tx_decoded_function_params'] = exactInputSingleTransactions['Tx_decoded_function_params'].apply(lambda x: x.strip("[]").split(","))
exactInputSingleTransactions
```

```python
exactInputSingleTransactions['tokenIn'] = exactInputSingleTransactions.apply(lambda row: row.Tx_decoded_function_params[0], axis = 1)
exactInputSingleTransactions['tokenOut'] = exactInputSingleTransactions.apply(lambda row: row.Tx_decoded_function_params[1], axis = 1)
exactInputSingleTransactions['fee'] = exactInputSingleTransactions.apply(lambda row: row.Tx_decoded_function_params[2], axis = 1)
exactInputSingleTransactions['recipient'] = exactInputSingleTransactions.apply(lambda row: row.Tx_decoded_function_params[3], axis = 1)
exactInputSingleTransactions['deadline'] = exactInputSingleTransactions.apply(lambda row: row.Tx_decoded_function_params[4], axis = 1)
exactInputSingleTransactions['amountIn'] = exactInputSingleTransactions.apply(lambda row: row.Tx_decoded_function_params[5], axis = 1)
exactInputSingleTransactions['amountOutMinimum'] = exactInputSingleTransactions.apply(lambda row: row.Tx_decoded_function_params[6], axis = 1)
exactInputSingleTransactions['sqrtPriceLimitX96'] = exactInputSingleTransactions.apply(lambda row: row.Tx_decoded_function_params[7], axis = 1)
exactInputSingleTransactions
```

```python
exactInputSingleTransactions.groupby(['tokenIn', 'tokenOut']).size().reset_index(name='counts').sort_values(by=['counts'], ascending=False).head(15)
```
